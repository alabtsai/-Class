```python
res=requests.get('http://search.books.com.tw/search/query/key/python/cat/all')
```

```python
import pandas as pd 
books=pd.Series()
print(books)
for book in soup.select("img[class='itemcov']"):
    print(book['alt'])
    print(type(book['alt']))
    books = books.append(pd.Series(book['alt']))
```
