```python
df1=pd.read_csv('REVIEW_x4qV0.csv')
```

```python
know_to_value={'知道':1,'不知道':0}
```

```python
df1['知_或_不知']=df1['是否知道行政院環保署的「環保集點」活動'].map(know_to_value)
```

```python
df1.pivot_table(values='知_或_不知',index=['職業','生理性別'],columns='年齡',aggfunc='count')
```

```python
df1.pivot_table(values='知_或_不知',index=['職業','生理性別'],columns='年齡',aggfunc='sum')
```

```python
df1.pivot_table(values='知_或_不知',index=['職業','生理性別'],columns='年齡',aggfunc='mean')
```
