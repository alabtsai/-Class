## 將建築完成年月 numpy.float64 轉成 datetime 

```python

from datetime import datetime
# 將 建築完成年月  轉成 西元 放入  build_date list

build_date=[]
for i in range(len(dfmo1)):
    quot,rem=divmod(dfmo1.loc[i,'建築完成年月'],100)
    yr1=int(quot+1911)
    mo1=int(rem)
    build_date.append(datetime(yr1,mo1,1))

dfmo1['建築完成 datetime']=build_date
```

## In the LinearRegression method in sklearn, what exactly is the fit_intercept parameter doing?

```python
from sklearn.linear_model import LinearRegression

# 產生 linear regression 物件
lin_reg=LinearRegression()
print(type(lin_reg))
print(lin_reg)

# 訓練模型
lin_reg.fit(X_train,y_train)
print('各變項參數:',lin_reg.coef_)

some_data=X_train[:5]
print(some_data)
some_label=y_train[:5]
print(some_label)
```
